```python
## lets conduct a comprehensive check to gather more information about our data
def data_quality_check(df_1):
    info_report = {
        'shape': df_1.shape,
        'total_records': len(df_1),
        'missing_records': df_1.isnull().sum(),
        'duplicate_rows': df_1.duplicated().sum(),
    }
    return info_report

info_report = data_quality_check(df_1)
print(info_report)
    ## RESULTS: Data has 3384 rows and 25 columns, Total records are 3384, There are NO Duplicate rows. However there are NULLS in some rows in respect to the column names displayed.
```

## Dealing with Nulls
- We can use the fillna() method to replace Nulls with some values. Case examples use either average records for Numeric Data, while Non Numeric data can be filled with 'Unknown'. This is important to ensure data analysis findings try reflect data

```python
def description_of_null_columns(df_1):
    desc_report = {
        'avg_age': df_1['age'].mean(),
        'avg_birth_year': df_1['birth_year'].mean(),
        'avg_wealth_gdp': df_1['wealth_gdp_ratio_pct'].mean(),
        'avg_wealt_citizen': df_1['wealth_per_citizen_usd'].mean(),
}
    return desc_report
desc_report = description_of_null_columns(df_1)
print(desc_report)

## The line of code is used to obtain Avg Values for the Numeric columns which then can be filled to records with Nulls
```

```python
def filled_null_records(df_1):
    fill_nulls = {
       'age': df_1['age'].fillna(65.0, inplace=True),
        'birth_year' : df_1['birth_year'].fillna(1960.0, inplace=True),
        'wealth_per_citizen_usd' : df_1['wealth_per_citizen_usd'].fillna(297.6, inplace=True),
        'wealth_gdp_ratio_pct' : df_1['wealth_gdp_ratio_pct'].fillna(0.8, inplace=True),
}
    return fill_nulls
fill_nulls = filled_null_records(df_1)
print(fill_nulls)

        
```

### Q4. Create an age bracket for the billionaires and group the billionaires count according  to their respective age brackets and split by gender 

```python
## Lets first inspect for the Maximum and Minimum Billionaire Ages
df_1['age'].max()
    # Maximum billionaire age is 104 years
```

```python
df_1['age'].min()
    # The Minimum billionaire age is 20 years
```

```python
## lets group the Billionaires into resoective age bracket
    #20-35 Years = Young enterpreneurs
    #36-55 Years = Established professionals
    #56-70 Years = Senior Wealth holders
    # 71+ Years = Old billionaires

## Define a function for Age Brackets
def age_bracket(age):
    if 20 <= age <= 35:
        return 'Young entrepreneurs'
    elif 36 <= age <= 55:
        return 'Established professionals'
    elif 56 <= age <= 70:
        return 'Senior Wealth holders'
    elif age >= 71:
        return 'Old billionaires'

df_1['age_bracket'] = df_1['age'].apply(age_bracket)
```

```python
## To view the Top 10 Rows
df_1.head(10)
```

```python
## We can then group the bilionaires by age age bracket split by gender. 
billionaire_stats = df_1.groupby(['age_bracket', 'gender']).agg(
    billionaire_count = ('person_name', 'count'),
    total_net_worth_m = ('net_worth_usd_m', 'sum')).reset_index()
billionaire_stats
```

```python
## To view age bracket with the most total_net_worth_m, we order in descENDING ORDER
billionaire_stats_order = billionaire_stats.sort_values('total_net_worth_m', ascending=False)
billionaire_stats_order
```

```python
## Which countries show a Monopoly in A Single industry with 100% absolute monopoly dominance?
    ## First find Unique industries per each country
country_indus = df_1.groupby('country')['primary_industry'].nunique().reset_index()
country_indus
```

```python
## Then obtain specific columns
country_indus_columns = ['country', 'primary_industry']
```

```python
##Filter to only countries only having ONLY one  Industry
total_monopoly = country_indus[country_indus['primary_industry']==1]
total_monopoly
```

```python
## Next we obtain Country names with Total Monopoly on an Industry
monopoly_stats = []
for country in total_monopoly['country']:
    industry = df_1[df_1['country']== country]['primary_industry'].iloc[0]
    billionaire_count = len(df_1[df_1['country'] == country])
    total_net_worth =  df_1[df_1['country'] == country]['net_worth_usd_m'].sum()

    monopoly_stats.append({
        'country': country,
        'monopoly_industry': industry,
        'billionaire_count': billionaire_count,
        'total_net_worth_usd_m': total_net_worth
    })

monopoly_df = pd.DataFrame(monopoly_stats).sort_values('total_net_worth_usd_m', ascending = False)
monopoly_df
                            

```

### Q6. Create filtered comparison for USA VS China Billionaires whose are Only Self-Made and the Primary indusry is Technology or Diversified. Which of the group has a higher wealth_gdp_ratio_pct?

```python
country_comparison= df_1[
    (df_1['country'].isin(['United States','China'])) &
    (df_1['self_made']== True) &
    (df_1['primary_industry'].isin(['Technology', 'Diversified']))
    ]

# Group by country and calculate statistics

comparison_stats = country_comparison.groupby('country').agg(
    avg_wealth_gdp_ratio=('wealth_gdp_ratio_pct', 'mean'),
    median_wealth_gdp_ratio=('wealth_gdp_ratio_pct', 'median'),
    count=('wealth_gdp_ratio_pct', 'count'),
    min_ratio=('wealth_gdp_ratio_pct', 'min'),
    max_ratio=('wealth_gdp_ratio_pct', 'max')
).reset_index()

print("=== USA vs CHINA: Self-Made Billionaires in Technology/Diversified ===")
print(comparison_stats)

## Result: United States of America has a higher average wealth to gdp ratio at 0.056 as compared to China's 0.029. The United States of America also has the Most Self Made Billionaires Billionaires at 212 while China has only 100
    
    
```
